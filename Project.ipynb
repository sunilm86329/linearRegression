#We can predict the weight based on height and weight dataset using linearRegression#

```python


import pandas as pd
dataset=pd.read_csv("/content/height-and-weight.csv")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, 1].values
dataset.isnull().sum() 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
#  Fit Linear Regression Model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train) 
# Predict values for test
lin_pred = lin_reg.predict(X_test) 
#  Compare predictions with real results
from sklearn import metrics
print('R square = ',metrics.r2_score(y_test, lin_pred))
print('Mean squared Error = ',metrics.mean_squared_error(y_test, lin_pred))
```

```python
#  Visualize Training set
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color="red")
plt.scatter(X_train, lin_reg.predict(X_train), color="blue" )
plt.title("Height and Weight - Training Set")

plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
#  Visualize Test set

import matplotlib.pyplot as plt
plt.scatter(X_test, y_test, color="red")
plt.scatter(X_test, lin_reg.predict(X_test), color="blue" )
plt.title("Height and Weight - Test Set")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

# Make a new prediction #

```python
# Make new Prediction
new_height=float(input(print("enter new height") )) 

lin_pred_new = lin_reg.predict([[new_height]])

print('If a person has height {},the predicted weight is {} '.format(new_height,lin_pred_new))
```
